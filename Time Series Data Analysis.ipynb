```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate date range with extra days for the rolling window
start_date = '2024-01-01'
end_date = '2024-12-31'
rolling_window_range = 30  # Number of extra days for the rolling window
dates = pd.date_range(start=start_date, end=end_date, freq='D')

# Generate random time series data
np.random.seed(0)  # for reproducibility
values = np.random.randn(len(dates))  # random values

# Create a DataFrame
data = pd.DataFrame({'Date': dates, 'Value': values})

# Plot the time series data
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Value'], color='blue', marker='o', linestyle='-')
plt.title('Sample Time Series Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.show()

# Basic statistics and information about the time series data
print("Basic Statistics:")
print(data.describe())

# Resampling the data (e.g., converting from daily to monthly frequency)
monthly_data = data.resample('M', on='Date').mean()
print("\nMonthly Data:")
print(monthly_data.head())

# Rolling window analysis with NaN values filled
rolling_avg = data['Value'].rolling(window=rolling_window_range).mean().fillna(method='bfill')

# Print rolling average values
print("\nRolling Window Analysis ({}-day Moving Average):".format(rolling_window_range))
print(rolling_avg.head())
```
